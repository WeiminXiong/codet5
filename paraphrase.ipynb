```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-xl")

model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-xl")

model.to("cuda:0")
```

```python
input = """please paraphrase the following sentence: \""" Input to this function is a string containing multiple groups of nested parentheses. Your goal is to separate those group into separate strings and return the list of those. Separate groups are balanced (each open brace is properly closed) and not nested within each other Ignore any spaces in the input string.\""""""
```

```python
inputs = tokenizer(input, return_tensors="pt").to('cuda:0')
```

```python
outputs = model.generate(inputs['input_ids'], max_length=200, do_sample = True, top_p=0.9, early_stopping=True)
```

```python
sentences = tokenizer.batch_decode(outputs, skip_special_tokens=True)
```

```python


from parrot import Parrot
import torch
import warnings
warnings.filterwarnings("ignore")


#Init models (make sure you init ONLY once if you integrate this to your code)
parrot = Parrot(model_tag="prithivida/parrot_paraphraser_on_T5")

```

```python
phrases = ["E.g. (()()) has maximum two levels of nesting while ((())) has three."]

def random_state(seed):
  torch.manual_seed(seed)
  if torch.cuda.is_available():
    torch.cuda.manual_seed_all(seed)

random_state(1234)

for phrase in phrases:
  print("-"*100)
  print("Input_phrase: ", phrase)
  print("-"*100)
  para_phrases = parrot.augment(input_phrase=phrase, use_gpu=False, do_diverse=True, diversity_ranker="levenshtein", max_return_phrases=10)
  for para_phrase in para_phrases:
    print(para_phrase)
```
